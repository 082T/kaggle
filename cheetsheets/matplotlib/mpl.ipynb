```python
import math
plt.plot([1,2,1,2,3],[1,3,4,1,2])
plt.show()
```

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
plt.plot(x,y) # connect (x,y) dots
plt.xlabel("input") # label x
plt.ylabel("sin") # label y
plt.title('sine wave') # plot title
plt.show()
```

```python
x = np.arange(0, math.pi*2, 0.2)
y = np.sin(x)

plt.plot(x,y, 'r.') # red .
plt.plot(x+1,y, 'g--') # green - 
plt.plot(x+2,y, 'b-.') # blue 2

plt.xlabel("input") 
plt.ylabel("sin") 
plt.title('sine wave') 
plt.show()
```

# Figure() - Object oriented interface

```python
fig = plt.figure() #empty canvas
ax = fig.add_axes([0,0,1,1]) # add axes: left, bottom, width, height

ax.set_title("sine wave")
ax.set_xlabel('input')
ax.set_ylabel('sine')

ax.plot(x,y)

plt.show()
```

# plt.subplots(a,b)

```python
x = np.linspace(0.1, 5, num = 20)

fig, a = plt.subplots(2,2)

a[0][0].plot(x, x*x)
a[0][1].plot(x, x)
a[1][0].plot(x, np.sqrt(x))
a[1][1].plot(x, np.log10(x))

a[0][0].set_title("x^2")
a[0][1].set_title("x")
a[1][0].set_title("sqrt(x)")
a[1][1].set_title("log10(x)")

a[0][0].axes.xaxis.set_ticks([])
a[0][0].axes.yaxis.set_ticks([])

fig.tight_layout() # da se razdvoje

plt.show()
```

### More freedom

```python
x = np.arange(1,12)

fig = plt.figure()

a1 = plt.subplot2grid((3,3),(0,0),colspan=2) # shape, location, rowspan, colspan
a2 = plt.subplot2grid((3,3),(0,2),rowspan=3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan=2,colspan=2)

a1.plot(x,x*x*x)
a2.plot(x,np.sin(x))
a3.plot(x,np.cos(x))

a3.set_title("cos")

fig.tight_layout()
plt.show()
```

# Custom grid

```python
x = np.arange(1,11)

fig, axes = plt.subplots(1,3, figsize=(16,5))

axes[0].plot(x,x**3, 'g', lw=9) # line width
axes[0].grid(True)
axes[0].set_title('Default grid')

axes[1].plot(x,np.exp(x),'r.')
axes[1].grid(color='b',ls='-.', lw=0.2)
axes[1].set_title('Custom grid')

axes[2].plot(x,x)
axes[2].set_title('No grid')

fig.tight_layout()
plt.show()
```

# Formatting axes

```python
x = np.arange(1,5)

fig,axes = plt.subplots(1,2, figsize=(10,4))

axes[0].plot(x,np.exp(x))
axes[0].plot(x, x**2)
axes[0].set_title("Normal scale")

axes[1].plot(x, np.exp(x))
axes[1].plot(x, x**2)
axes[1].set_yscale("log") # scale y
axes[1].set_title("Log scale y")

axes[0].set_xlabel("x axis")
axes[0].set_ylabel("y axis")
axes[0].xaxis.labelpad = 20 # label distance from graph

axes[1].set_xlabel("x axis")
axes[1].set_ylabel("y axis")

plt.show()
```

# Contour plot

```python
xlist = np.linspace(-3,3,100)
ylist = np.linspace(-3,3,100)

x, y = np.meshgrid(xlist,ylist) # matrix from -3 to 3 values 
z = np.sqrt(x**2 + y**2)

fig, ax = plt.subplots(1,1)

# contour, x,y = cartesian grid, z = function, levels = levels of different color
cp = ax.contourf(x,y,z, levels = np.linspace(0,5,200)) 

fig.colorbar(cp) # add colorbar
ax.set_title('Filled Contours Plot')
ax.set_ylabel('y cm')
ax.set_xlabel('x cm')

plt.show()
```

# Quiver plot

```python
x, y = np.meshgrid(np.arange(-2,2,.2), np.arange(-2,2,.25)) # grid
z = x*np.exp(-x**2 - y**2) # third dimension
v, u = np.gradient(z, .2, .2) # 

fig, ax = plt.subplots()
ax.quiver(x,y,u,v) # coords, gradient

plt.show()
```

# 3D plotting

```python
from mpl_toolkits import mplot3d
```

```python
z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot3D(x,y,z,'red') #plot from sets of triples
ax.set_title('3d plot')

plt.show()
```

# 3d scatter plot 

```python
z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)

c = x+y

fig = plt.figure()
ax = plt.axes(projection='3d')

t = ax.scatter(x,y,z, c = c)
ax.set_title('3d scatter plot')

fig.colorbar(t)

plt.show()
```

# 3d contour 

```python
a = np.linspace(-6,6,100)
b = np.linspace(-6,6,100)

x, y = np.meshgrid(a,b)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()

ax = plt.axes(projection='3d')

t = ax.contour3D(x, y, z, 100, cmap='binary') # gridx, gridy, function z, number of step details, color

fig.colorbar(t)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.set_title('3d contour')

plt.show()
```

# 3d wireframe

```python
a = np.linspace(-6,6,100)
b = np.linspace(-6,6,100)

x, y = np.meshgrid(a,b)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()

ax = plt.axes(projection='3d')

ax.plot_wireframe(x, y, z, color = 'black') # gridx, gridy, function z, number of step details, color

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.set_title('3d wireframe')

plt.show()
```

# 3d surface plot

```python
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T # transpose
z = np.cos(x ** 2 + y ** 2)

fig = plt.figure()
ax = plt.axes(projection='3d')

t = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
fig.colorbar(t)

ax.set_title('Surface plot')

plt.show()
```

# Text

text, annotate, xlabel, ylabel, title, figtext, suptitle

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.set_title('Axes title')
plt.suptitle('Figure suptitle')

ax.set_xlabel('Axes xlabel')
ax.set_ylabel('Axes ylabel')

ax.text(7, 7, 'random text', style = 'italic', bbox = { 'facecolor': 'red' }, fontsize = 20)
ax.text(5, 2, r'an equation: $E=mc^2$', fontsize = 15)
ax.text(3, 6, 'Colored text in axes coords', verticalalignment='bottom', color='green', fontsize=15)

ax.plot(2, 1, 'o')

ax.annotate('annotate', xy = (2, 1), xytext = (3, 4), arrowprops = dict(facecolor = 'blue', shrink = 0.1) )

ax.axis([0,10,0,10])
plt.show()
```
