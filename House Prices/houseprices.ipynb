```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
sample_submission = pd.read_csv('sample_submission.csv')

print(f"Train data length: {len(train_data)}, test data length: {len(test_data)}")
```

```python
with open('data_description.txt') as f:
    contents = f.read()
    print(contents)
```

```python
prices = train_data['SalePrice']
log_scaled_prices = np.log(prices)
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,3))
ax.hist(prices, bins = 100, color = "orange",ec = "black")
plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,3))
ax.hist(log_scaled_prices, bins = 100, color = "orange",ec = "black")
plt.show()
```

```python
train_data = train_data.drop('SalePrice', axis=1)
full_data = train_data.append(test_data).reset_index().drop('index', axis=1).drop('Id', axis=1)
full_data = pd.get_dummies(full_data)
```

```python
columns_with_nan_values = []

for column in full_data:
    if full_data[column].isnull().any() == True:
        columns_with_nan_values.append(column)
        
for column in columns_with_nan_values:
    full_data[column] = full_data[column].fillna(full_data[column].median())
```

```python
to_train = full_data.loc[:1459]
to_test = full_data.loc[1460:]
```

```python
x_train, x_test, y_train, y_test = train_test_split(to_train, log_scaled_prices, test_size=0.2)
```

## SVR

```python
costs = [0.00003, 0.0003, 0.003, 0.03, 0.3, 3, 30]
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
divider = ""
```

```python
for kernel in kernels:
    for cost in costs:
        regr = make_pipeline(StandardScaler(), SVR(kernel=kernel, C=cost, epsilon=0.01))
        regr.fit(x_train, y_train)
        print(f"Kernel: {kernel}, cost: {cost}, regr.score: {round(regr.score(x_test, y_test), 4)}")
    print(divider)
```

```python
regr = make_pipeline(StandardScaler(), SVR(kernel='linear', C=0.0003, epsilon=0.01))
regr.fit(to_train, log_scaled_prices)
to_submit = regr.predict(to_test)
```

```python
to_submit = np.exp(to_submit)
```

```python
submit_df = sample_submission
submit_df['SalePrice'] = to_submit
```
